```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Results 3-fold Cross-validation**

```python
cv_results_hdbscan = pd.DataFrame(pd.read_csv("data/accuracy_hdbscan.csv").mean().rename("Accuracy"))
cv_results_kmeans = pd.DataFrame(pd.read_csv("data/accuracy_kmeans.csv").mean().rename("Accuracy"))
```

```python
print("Number of clusters of K-Means averaged over the folds:")
print(cv_results_kmeans)
best_accuracy_kmeans = cv_results_kmeans.max().values[0].item()
print("Optimal number of clusters: ")
print(cv_results_kmeans[cv_results_kmeans["Accuracy"]==best_accuracy_kmeans])
```

```python
print("Number of clusters of HDBscan averaged over the folds:")
print(cv_results_hdbscan)
best_accuracy_hdbscan = cv_results_hdbscan.max().values[0].item()
print("Optimal number of clusters: ")
print(cv_results_hdbscan[cv_results_hdbscan["Accuracy"]==best_accuracy_hdbscan])
```

```python
plt.plot(cv_results_kmeans, label='KMeans', marker='o')

# Add labels, title, and legend
plt.xlabel('Number of Clusters')   # adjust as needed
plt.ylabel('Accuracy')           # adjust as needed
plt.title('Cross-Validation Results K-Means')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(cv_results_hdbscan, label='HDBScan', marker='o')

# Add labels, title, and legend
plt.xlabel('Minimal Cluster size')   # adjust as needed
plt.ylabel('Accuracy')           # adjust as needed
plt.title('Cross-Validation Results HDBscan')
plt.legend()
plt.grid(True)
plt.show()
```
